```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel('Python-Machine-Learning/Dataset/titanic3 (3).xls')
```

```python
data.dropna(axis=0, how='any')
data.head(10)
```

```python
data.drop(['name','sibsp','parch','ticket','fare','cabin','embarked','boat','body','home.dest'], axis = 1, inplace=True)
```

```python
data.head(10)
```

```python
data.describe()
```

```python
data['pclass'].value_counts().plot.bar()
```

```python
data['survived'].value_counts().plot.bar()
```

```python
data['age']
```

```python
data.loc[(data['age'] < 20), 'age'] = 0
data.loc[(data['age'] >= 20) & (data['age'] < 30), 'age'] = 1
data.loc[(data['age'] >= 30) & (data['age'] < 40), 'age'] = 2
data.loc[(data['age'] > 40), 'age'] = 3
```
